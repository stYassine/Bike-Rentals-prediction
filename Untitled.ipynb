# PROBLEM STATEMENT

#### - The Objective is to predict bike rental usage based on inputs such as temperature, humidity, wind speed, ...etc.
#### - This problem is is more advanced, so we have to build a multi-layer perceptron network.

#### - Inputs:
	- Instant: record index
	- dteday: date
	- season: season (1: spring, 2: summer, 3: fall, 4, winter)
	- yr: year (0:2011, 1:2012)
	- mnth: month (1 to 12)
	- hr: hour (0 to 23)
	- holiday: whether day is holiday or not - 
	- weekday: day of the week
	- working day: if day is neither weekend nor holiday is 1, otherwise is 0
    - weathersit:
        1: Clear, Few clouds, Partly Cloudy
        2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, mist
        3: Light snow, light rain + Thunderstorm + Scatterd clouds, Light Rain + Scatterd clouds
        4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
    - temp: Normalized temperature in Celsius. The values are divided to 41 (max)
    - windspeed: Normalized wind speed, The values are divided to 67 (max)

#### - Outputs:
    - casual: count of casual users
    - registered: count of registered users
    - cnt: count of total rental bikes including both casual and registerd

# 4- CREATE TRAINING AND TESTING DATASET

```python
X_cat = bike_df[['season', 'yr', 'mnth', 'holiday', 'weekday',
       'workingday', 'weathersit', ]]
```

```python
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
oneHotEncoder = OneHotEncoder()
X_cat = oneHotEncoder.fit_transform(X_cat).toarray()
```

```python
X_cat
```

```python
X_cat.shape
```

```python
X_cat = pd.DataFrame(X_cat)
```

```python
X_cat
```

```python
X_numerical = X_numerical.reset_index()
```

```python
X_numerical
```

```python
X_all = pd.concat([X_cat, X_numerical] , axis = 1)
```

```python
X_all
```

```python
X_all = X_all.drop(['dteday'], axis = 1)
```

```python
X = X_all.iloc[:, :-1].values
y = X_all.iloc[:, -1:].values
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
y = scaler.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

# 5- TRAIN THE MODEL

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(units=100, activation='relu', input_shape = (35,) ))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=1, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss='mean_squared_error')
```

```python
epochs = model.fit(X_train, y_train, epochs=25, batch_size=50, validation_split = 0.2)
```

# 6- EVALUATE THE MODEL

```python
epochs.history.keys()
```

```python
plt.plot(epochs.history['loss'])
plt.plot(epochs.history['val_loss'])

plt.title('Model Loss Progress During Training')
plt.xlabel('Epochs')
plt.ylabel('Training Loss')
```

```python
y_predict = model.predict(X_test)
plt.plot(y_test, y_predict, '^', color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```

```python
y_predict_origin = scaler.inverse_transform(y_predict)
```

```python
y_test_origin = scaler.inverse_transform(y_test)
```

```python
plt.plot(y_test_origin, y_predict_origin, '^', color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```

```python
K = X_test.shape[1]
```

```python
K
```

```python
n = len(X_test)
```

```python
n
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt
```

```python
RMSE = float(format(np.sqrt(mean_squared_error(y_test_origin, y_predict_origin)), '0.3f'))
MSE = mean_squared_error(y_test_origin, y_predict_origin)
MAE = mean_absolute_error(y_test_origin, y_predict_origin)
r2 = r2_score(y_test_origin, y_predict_origin)
adj_r2 = 1-(1-r2)*(n-1)/(n-K-1)
```

```python
print('RMSE = ', RMSE)
print('MSE = ', MSE)
print('MAE = ', MAE)
print('R2 = ', r2)
print('Adjusted R2 = ', adj_r2)
```
